```python
#Programme python pour convertir les fichiers .json en .csv 
import pandas as pd
import json
import csv
#ouverture du fichier json et importation des données dans la variable data
with open('mpd.slice.0-999.json') as json_file:
    data = json.load(json_file)
df = pd.json_normalize(data['playlists'])
df.to_csv("playlist.csv")

#Affichage des données
df
```

```python
#Indexion du numéro de playlist par morceau
f=0
for i in range(0,len(df["tracks"])):
    for j in range(0,len(df["tracks"][i])):
        x=df["tracks"][i][j]
        x["pid"]=f

    f+=1
```

```python
#Programme python pour convertir les fichiers .json en .csv 
import json
import csv
#ouverture d'un fichier .csv pour écrire dessus
music_list = open('morceaux.csv', 'w')
#Création d'un objet d'écriture pour python
csv_writer = csv.writer(music_list)
# Variable de comptage utilisée pour écrire les en-têtes dans le fichier CSV
count = 0
for i in range(0,len(df["tracks"])):
    for j in range(0,len(df["tracks"][i])):
            x=df["tracks"][i][j]
            # Écriture des en-têtes du fichier CSV
            if count==0:  
                header = x.keys()
                csv_writer.writerow(header)
                count+=1
            csv_writer.writerow(x.values())
            
        # Écriture des données du fichier CSV            
music_list.close()
```

```python
#Ouverture de morceaux.csv et affichage de ce dernier
df_music = pd.read_csv("morceaux.csv")
df_music.head(10)
```

```python
#Définition de df_liaison comme d'un dataframe
df_liaison=pd.DataFrame()
#Définition de la colonne Tracks de liaison comme la colonne Track_uri de music
df_liaison["tracks"]=df_music["track_uri"]
#Création d'une colonne ID vide
df_liaison["ID"]=""

from collections import defaultdict
neo = df_liaison["tracks"]
d = defaultdict(lambda: len(d))  
neo_id = [d[x] for x in neo]
df_liaison["ID"]=neo_id
```

```python
#Création de deux colonnes vides "start_id(track)" et "end_id(track)"
df_liaison[":start_id(track)"]=""
df_liaison[":end_id(track)"]=""
```

```python
#Remplissage des colonnes "start_id" et "end_id"
for i in range(0,len(df_liaison["tracks"])):
    df_liaison[":start_id(track)"][i]=df_liaison["ID"][i]
    if i==67502:
        df_liaison[":end_id(track)"][i]=7
    else:
        df_liaison[":end_id(track)"][i]=df_liaison["ID"][i+1]
```

```python
#Suppression de la colonne tracks
df_liaison=df_liaison.drop(columns="tracks")
#Suppression de la colonne ID
df_liaison=df_liaison.drop(columns="ID")
```

```python
#Ajout de la colonne duration qui vaut la colonne duration_ms de df_music
df_liaison["duration"]=df_music["duration_ms"]
```

```python
#Création d'un .csv et ajout des données dedans
df_liaison.to_csv('liaison.csv',index=False)
```

```python
#Importation de GraphDatabase via neo4j
from neo4j import GraphDatabase

#Connexion à neo4j
uri = "bolt://localhost:7687"
driver = GraphDatabase.driver(uri, auth=("neo4j", "esme"))
```

```python
#Affichage de tous les morceaux d'un artiste choisi
def get_singers(tx,name_artist):
    query = """MATCH (n:Morceaux) where n.artist_name=$name_artist return n.track_name as Titre ORDER BY n.track_name"""
    result = tx.run(query,name_artist=name_artist)
    return result.data()

with driver.session(database="tpfinal") as session:
    singers = session.read_transaction(get_singers, "Beyoncé")
    for a in singers:
        print(a)
```

```python
#Affichage du chemin le plus court entre 2 morceaux
with driver.session(database="tpfinal") as session:
        paths = session.run("""MATCH (n:Morceaux{track_name: "Stairway To Heaven"}), (m:Morceaux{track_name: "Dangerous (feat. Sam Martin)"})
MATCH p=shortestPath((n)-[a:Liaison*]->(m)) RETURN p""")
        chemin=paths.data()
        print(chemin)
```
